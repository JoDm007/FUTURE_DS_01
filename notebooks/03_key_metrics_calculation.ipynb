1. Importations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import os

# path
os.chdir('e:/Certifs/Future_Interns/FUTURE_DS_01')
```

```python
df = pd.read_csv("data/processed/cleaned_superstore.csv", parse_dates=['Order Date'])
```

2. Globals metrics

```python
total_revenue = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_margin = (total_profit / total_revenue) * 100
total_quantity = df['Quantity'].sum()
avg_discount = df['Discount'].mean() * 100

print("=" * 50)
print("📊 MÉTRIQUES GLOBALES")
print("=" * 50)
print(f"💰 Chiffre d'affaires : ${total_revenue:,.2f}")
print(f"📈 Profit total : ${total_profit:,.2f}")
print(f"🎯 Marge bénéficiaire : {avg_margin:.2f}%")
print(f"📦 Quantité vendue : {total_quantity:,.2f}")
print(f"🏷️ Discount moyen : {avg_discount:.2f}%")
print("=" * 50)
```

3. Top products

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
print("\n🏆 TOP 10 PRODUITS PAR CA")
print(top_products.to_string(index=False))
```

4. Categories

```python
category_analysis = df.groupby('Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).sort_values('Sales', ascending=False).reset_index()

category_analysis['Margin'] = (category_analysis['Profit'] / category_analysis['Sales']) * 100
print("\n ANALYSE PAR CATEGORIE")
print(category_analysis.to_string(index=False))
```

5. Regions

```python
region_analysis = df.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).sort_values('Sales', ascending=False).reset_index()

region_analysis['Margin'] = (region_analysis['Profit'] / region_analysis['Sales']) * 100
print("\n🌍 ANALYSE PAR RÉGION")
print(region_analysis.to_string(index=False))
```

6. Tendance mensuelle

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().reset_index()
monthly_sales['Order Date'] = monthly_sales['Order Date'].astype(str)
print("\n📅 TOP 5 MOIS AVEC LE PLUS DE VENTES")
print(monthly_sales.sort_values('Sales', ascending=False).head().to_string(index=False))
```
